```python
import pandas as pd
#Loading Dataset
data = pd.read_csv("kurtiData.csv")
```

```python
#Appending producct url to a list
col=[]
for i, row in data.iterrows():
  col.append(row[2])
```

```python
data['product_url']
```

```python
#Fetching product name from product url
import re
prodnames=[]
pattern = r"/([^/]+)/p/"
for i in range(len(col)):
  match = re.search(pattern, col[i])
  if match:
    extracted_string =match.group(1)
    prodnames.append(extracted_string)
```

```python
# adding prodnames as an extra column to the data
data['product_name']=prodnames
data.head()
```

```python
productRatings = data.pivot_table(index=['user_id'],columns=['product_name'],values='rating')
productRatings.head()
```

```python
kurtisRatings = productRatings['aagam-drishya-kurtis']
kurtisRatings.head()
```

```python
similarKurtis = productRatings.corrwith(kurtisRatings)
similarKurtis = similarKurtis.dropna()
df = pd.DataFrame(similarKurtis)
df.head(10)
```

```python
similarKurtis.sort_values(ascending=False)
```

```python
import numpy as np
productStats = data.groupby('product_name').agg({'rating': [np.size, np.mean]})
productStats.head()
```

```python
popularKurtis = productStats['rating']['size'] >= 50
productStats[popularKurtis].sort_values([('rating', 'mean')], ascending=False)[:15]
```

```python
#df = movieStats[popularMovies].join(pd.DataFrame(similarMovies, columns=['similarity']))

# Updated for newer Pandas releases that don't allow merging between different levels; we must flatten it first now.
mappedColumnsproductStats=productStats[popularKurtis]
mappedColumnsproductStats.columns=[f'{i}|{j}' if j != '' else f'{i}' for i,j in mappedColumnsproductStats.columns]
df = mappedColumnsproductStats.join(pd.DataFrame(similarKurtis, columns=['similarity']))
```
