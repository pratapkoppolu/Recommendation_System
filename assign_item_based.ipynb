## Item-based collaborative filtering

```python
import pandas as pd
#Loading Dataset
data = pd.read_csv("kurtiData.csv")
```

```python
data.head()
```

```python
#Appending producct url to a list
col=[]
for i, row in data.iterrows():
  col.append(row[2])
```

```python
data['product_url']
```

```python
#Fetching product name from product url
import re
prodnames=[]
pattern = r"/([^/]+)/p/"
for i in range(len(col)):
  match = re.search(pattern, col[i])
  if match:
    extracted_string =match.group(1)
    prodnames.append(extracted_string)
```

```python
# adding prodnames as an extra column to the data
data['product_name']=prodnames
data.head()
```

```python
userRatings = data.pivot_table(index=['user_id'],columns=['product_name'],values='rating')
userRatings.head()
```

```python
userRatings.shape
```

```python
corrMatrix = userRatings.corr()
corrMatrix.head()
```

```python
corrMatrix = userRatings.corr(method='pearson', min_periods=100)
corrMatrix.head()
```

```python
userRatings.reset_index(drop=True, inplace=True)
myRatings = userRatings.loc[0].dropna()
print(myRatings)
```

```python
simCandidates = pd.Series()
for i in range(0, len(myRatings.index)):
    print ("Adding sims for " + myRatings.index[i] + "...")
    # Retrieve similar products to this one that I rated
    sims = corrMatrix[myRatings.index[i]].dropna()
    # Now scale its similarity by how well I rated this kurti
    sims = sims.map(lambda x: x * myRatings[i])
    # Add the score to the list of similarity candidates
    simCandidates = pd.concat([simCandidates, sims])
    
#Glance at our results so far:
print ("sorting...")
simCandidates.sort_values(inplace = True, ascending = False)
print (simCandidates.head(10))
```

```python
simCandidates = simCandidates.groupby(simCandidates.index).sum()
```

```python
simCandidates.sort_values(inplace = True, ascending = False)
simCandidates.head(10)
```

The last thing we have to do is filter out movies I've already rated, as recommending a movie I've already watched isn't helpful:

```python
filteredSims = simCandidates.drop(myRatings.index)
filteredSims.head(20)
```
