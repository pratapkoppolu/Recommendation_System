## Data Analysis

```python
import pandas as pd
#Loading Dataset
data = pd.read_csv("kurtiData.csv")
```

```python
#data.head()
```

```python
#Appending producct url to a list
col=[]
for i, row in data.iterrows():
  col.append(row[2])
```

```python
data['product_url']
```

```python
#Fetching product name from product url
import re
prodnames=[]
pattern = r"/([^/]+)/p/"
for i in range(len(col)):
  match = re.search(pattern, col[i])
  if match:
    extracted_string =match.group(1)
    prodnames.append(extracted_string)
```

```python
# adding prodnames as an extra column to the data
data['product_name']=prodnames
data.head()
```

```python
#using popularity index to get  products based on ratings
num_rating =  data.groupby('product_name').count()['rating'].reset_index()
num_rating.rename(columns={'rating':'num_ratings'},inplace=True)
num_rating
```

```python
#using popularity index to get  products based on average rating
avg_rating =  data.groupby('product_name').mean()['rating'].reset_index()
avg_rating.rename(columns={'rating':'avg_ratings'},inplace=True)
avg_rating
```

```python
# merging products based num_ratings and avg_ratings
popular = num_rating.merge(avg_rating,on='product_name')
popular
```

```python
popular[popular['num_ratings']>=100]
```

```python
#Displaying popular products 
popular_sorted = popular[popular['num_ratings'] >= 150].sort_values(by='avg_ratings',ascending = False).head(50)
popular_sorted['avg_ratings'] = round(popular_sorted['avg_ratings'], 2)
popular_products = popular_sorted.merge(data,on='product_name').drop_duplicates('product_name')[['product_name','product_id','image_url','price','num_ratings','avg_ratings']]
popular_products
```

```python
#setting threshold for users who've rated more than 10 products
x=data.groupby('user_id').count()['rating'] > 10
valuable_user = x[x].index
valuable_user
```

```python
user_ratings = data[data['user_id'].isin(valuable_user)]
```

```python
#setting threshhold for product names having more than 300 ratings
y = user_ratings.groupby('product_name').count()['rating'] >= 150
popularProductname=y[y].index
popularProductname
```

```python
final_ratings = user_ratings[user_ratings['product_name'].isin(popularProductname)]
```
